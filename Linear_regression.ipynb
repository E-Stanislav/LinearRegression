```python
import numpy as np
```

```python
x_hw = np.array([[50],[60],[70],[100]])
x_hw
```

```python
y_hw = np.array([[10],[15],[40],[45]])
y_hw
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
margin = 10
```

```python
x_min = x_hw[:,0].min() - margin
x_max = x_hw[:,0].max() + margin
```

```python
plt.xlim(x_min, x_max)
plt.ylim(0, y_hw[:,0].max() + margin) 
plt.scatter(x_hw[:,0], y_hw[:,0], 40, 'g')
```

Нахождение коэффициентов для линейной регрессии

```python
from numpy.linalg import inv
```

```python
w = inv(
    (x_hw.T).dot(x_hw)
).dot(
    x_hw.T
).dot(y_hw)

w
```

```python
x_support = np.linspace(x_min, x_max, 100)
```

```python
y_model = w[0][0] * x_support
```

```python
y_pred = w[0][0] * x_hw
```

```python
 plt.scatter(x_hw[:,0], y_hw[:,0], 40, 'g')
plt.plot(x_support, y_model)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(x_hw,y_hw)
```

```python
pred_sk = reg.predict(x_hw)
```

Построение уровнения для линейной регрессии, после использования библиотеки sklearn 

```python
y_reg = x_support * reg.coef_[0][0] + reg.intercept_[0]
```

```python
plt.scatter(x_hw[:,0], y_hw[:,0], 40, 'g')
plt.plot(x_support, y_model, ':')
plt.plot(x_support, y_reg, '--')
plt.show()
```

Сравнение полученных коэффициентов детерминации для полученных вручную результатов и расчитанных с помощью библиотеки sklearn

```python
from sklearn.metrics import r2_score
```

Полученные вручную

```python
r2_score(y_hw, y_pred)
```

Полученные с помощью библиотеки sklearn

```python
r2_score(y_hw, pred_sk)
```
