```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score
from matplotlib import pyplot as plt
from sklearn.linear_model import SGDRegressor
from scipy.spatial import distance
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('3.10_non_linear.csv')
plt.scatter(df.x_train, df.y_train)
```

```python
x_train, x_value, y_train, y_value = train_test_split(df.x_train.values.reshape(-1,1), df.y_train.values, test_size = 0.2, random_state = 42)
sgd_regress = SGDRegressor(learning_rate='constant', eta0=0.01, fit_intercept=True, random_state=42)
```

```python
sgd_regress.partial_fit(x_train, y_train)
x_linspace = np.linspace(df.x_train.min(), df.x_train.max(), num =100)
y_linspace = sgd_regress.predict(x_linspace.reshape(-1, 1))
```

```python
plt.scatter(x_train, y_train, 30, 'g')
plt.scatter(x_value, y_value, 70, 'r', '*')
plt.plot(x_linspace, y_linspace)
```

```python
y_value_pred = sgd_regress.predict(x_value.reshape(-1,1))
```

```python
r2_score(y_value, y_value_pred)
```
