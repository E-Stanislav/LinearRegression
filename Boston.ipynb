```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
```

```python
boston_data = load_boston()
```

```python
df = pd.DataFrame(boston_data.data, columns = boston_data.feature_names)
df_target = pd.Series(boston_data.target)
```

Разделил на тренировочную и валидную часть


```python
train_df = df.head(405)
valid_df = df.tail(len(df)-405)
train_target_df = df_target.head(405)
valid_target_df = df_target.tail(len(df)-405)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
reg = LinearRegression().fit(train_df, train_target_df)
pred_1 = reg.predict(train_df)
```

Нахождение R2 (коэффициента детерминации) для тренировочной части

```python
r2_train = r2_score(train_target_df, pred_1)
r2_train
```

Z-score преобразование

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
trans_boston = MinMaxScaler().fit_transform(train_df)
```

```python
reg_2 = LinearRegression().fit(trans_boston, train_target_df)
pred_2 = reg_2.predict(trans_boston)
```

```python
r2_zscore = r2_score(train_target_df,pred_2)
r2_zscore
```

Сравнение коэффициентов детерминации

```python
r2_train < r2_zscore
```
